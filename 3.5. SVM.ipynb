## Introduction

A support vector machine (SVM) is a supervised machine learning algorithm that can be used for both classification and regression tasks. The algorithm is a discriminative classifier that maximizes the margin between the classes in the feature space. Support vector machines are effective in high dimensional spaces and are also capable of handling non-linear classification tasks. 

### Data Loading and Processing

As SVMs learn decision boundaries just like logistic regresssion models, we will have to process the numeric columns as we have previously, transforming certain features using BoxCox transformations.

## Modelling

### Baseline Model

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split

X, y = model_utils.GenerateXy(df)

X_rem, X_test, y_rem, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=1)

column_transformer = model_utils.CreateColumnTransformer(X)

X_rem_transformed = column_transformer.fit_transform(X_rem)
X_test_transformed = column_transformer.transform(X_test)

svm = LinearSVC()
svm.fit(X_rem_transformed, y_rem)

svm.score(X_test, y_test)
```

Alternatively, we can employ a kernel SVM.

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

linear_kernel_svm = SVC(kernel='linear')
linear_kernel_svm.fit(X_rem_transformed, y_rem)
```

```python
rbf_kernel_svm = SVC(kernel='rbf')
rbf_kernel_svm.fit(X_rem_transformed, y_rem)
```
